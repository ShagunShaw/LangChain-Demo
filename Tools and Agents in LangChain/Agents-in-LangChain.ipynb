```python
from langchain.openai import ChatOpenAI
from langchain.tools import tool
from langchain_community.tools import DuckDuckGoSearchRun
import requests
```

```python
llm= ChatOpenAI()
```

## Importing our Agents

```python
from langchain.agents import create_react_agent, AgentExecutor
from langchain import hub
```

```python
# Pulling the ReAct prompt from LangChain Hub
prompt= hub.pull("hwchase17/react")     # pulls the standard react agent prompt
```

```python
# Create the ReAct agent manually with the pulled prompt
agent= create_react_agent(
    llm= llm,
    tools= [search_tool],
    prompt= prompt
)
```

```python
# Wrap it with Agent Executer
agent_executer= AgentExecutor(
    agent= agent,
    tools= [search_tool],
    verbose= True     # 'verbose= True' means it also shows detailed logs of how the agent is working behind the scenes, instead of just printing the output
)
```

The difference between agent and agent_executer is that agent thinks what to be done next and based on that our agent_executer performs the specific task. For example, if agent thinks for a given problem he has to hit the weather api, then the agent_executer will hit the wearther api and fetches the required info and gives it back to the agent.

```python
# Invoke
response= agent_executer.invoke({"input": "find the capital of Madhya Pradesh, then find it's current weather condition"})

print(response)   # will print the logs along with the output
print(response['output'])     # will print the output only
```

## what is ReAct in LangChain?
ReAct = “Reasoning + Acting”
It's a method used by LLM agents in LangChain to decide what to think and which tool to use next.

What does ReAct do?

Ans) ReAct gives the LLM a pattern like:

Thought → model reasons about what to do
Action → model chooses a tool
Action Input → gives input to the tool
Observation → tool returns result
Repeat…
Final Answer

**Why is it useful?**
*   Helps the agent break down a task step-by-step
*   Avoids hallucinations
*   Lets the LLM justify why it uses certain tools
*   Makes tool use more reliable
